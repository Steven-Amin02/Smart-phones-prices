```python
numeric_cols = df[['battery_capacity', 'fast_charging_power']]

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # outlier condition
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    
    print(f"{col}: {len(outliers)} outliers")
```

```python
import numpy as np
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    
    print(f"{col}: {len(outliers)} outliers")
```

```python
numeric_cols = df[['battery_capacity', 'fast_charging_power','Resolution_Width']]

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df[col] = df[col].clip(lower=lower, upper=upper) 
```

```python
import pandas as pd
import numpy as np

def convert_to_gb(storage):
    if 'TB' in storage:
        return float(storage.replace('TB', '')) * 1000
    elif 'GB' in storage:
        return float(storage.replace('GB', ''))
    else:
        return np.nan

df['storage_gb'] = df['memory_card_size'].apply(convert_to_gb)
```

```python
df = df.drop(columns=['memory_card_size'])
```
